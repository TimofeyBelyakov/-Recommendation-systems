Заменим выбросы на средние значения по признаку:

```python
for i in range(0, len(data)):
    if data['dtnjw'][i] < -20:
        data['dtnjw'][i] = data['dtnjw'].mean()
data['dtnjw'].hist(bins=30)
pl.suptitle("dtnjw")
```

```python
for i in range(0, len(data)):
    if data['muldh'][i] < -25:
        data['muldh'][i] = data['muldh'].mean()
data['muldh'].hist(bins=30)
pl.suptitle("muldh")
```

```python
for i in range(0, len(data)):
    if data['djrge'][i] < -50:
        data['djrge'][i] = data['djrge'].mean()
data['djrge'].hist(bins=30)
pl.suptitle("djrge")
```

```python
print("Новые средние значения по столбцам:")
mean_new = data.mean()
print(mean_new)
print("\n")

print("Новые дисперсии по столбцам:")
disp_new = data.std() * data.std()
print(disp_new)
print("\n")
```

Произведём нормализацию:

```python
mean1 = data['dtnjw'].mean()
mean2 = data['muldh'].mean()
mean3 = data['djrge'].mean()
disp1 = data['dtnjw'].std() * data['dtnjw'].std()
disp2 = data['muldh'].std() * data['muldh'].std()
disp3 = data['djrge'].std() * data['djrge'].std()

for i in range(0, len(data)):
    data['dtnjw'][i] = (data['dtnjw'][i] - mean1) / disp1
    data['muldh'][i] = (data['muldh'][i] - mean2) / disp2
    data['djrge'][i] = (data['djrge'][i] - mean3) / disp3
```
